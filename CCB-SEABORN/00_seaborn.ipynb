```python
# Dataframe

import pandas as pd
import numpy as np
import seaborn as sns

path = r"C:\Work\CCB-PYTHON-DS"
file = "\\databasenew_world_00.csv"

# Import file
df = pd.read_csv(path+file)
df.columns
```

```python
df_subset_one = df[['idx','income','save_profile','sex','education']]

# Line and scatter
sns.lineplot(data= df_subset_one, x='idx',y='income')
```

```python
# Scatter plot
sns.scatterplot(data = df_subset_one,
                 x= 'save_profile', y='income',
                 hue = 'sex')
```

```python
# Bar
sns.barplot(data= df_subset_one, x='sex',y='save_profile',palette='deep')
```

```python
# Cat
result = df.groupby(["sex"])['income'].aggregate(np.median).reset_index().sort_values('income')
sns.barplot(data = df_subset_one, x= 'sex',y='income', hue='education') #hue
```

```python
# Histogram
sns.histplot(data=df_subset_one, x="income")
```

```python
# Boxplot
sns.boxplot(x="education", y="income",
            hue="sex", palette=["blue", "red"],
            data=df_subset_one)
```

```python
# Correl
df_corr = df[['education','sex','income','save_profile']].corr()
print(df_corr)

sns.heatmap(df_corr, annot=True, cmap = 'magma')
```
