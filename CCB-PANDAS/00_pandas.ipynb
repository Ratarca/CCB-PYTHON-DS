```python
import pandas as pd

path = r"C:\Work\CCB-PANDAS\database"
file = "\\world.csv"

# Import file
df = pd.read_csv(path+file)

df.head(n=5)
```

```python
"""
Columns
- Check Columns
- Create column
- Drop column
- Rename Column
- Operation with columns
"""

# Check column
print(f"Columns now are : {list(df.columns)}")

# Create Column
df['new_col'] = 0

# Drop Column
df = df.drop(columns='new_col')

# Rename Column
df = df.rename(columns = {'Unnamed: 0':'idx'})

# Operate with columns
df['percentage_save'] = df['save_profile']/df['income']
```

```python
"""
Statistics Dataframe
- Info and Describe
- Unique vals (nuquie)
- Metrics : sum /mean / std/
- Count values
"""

print(" ### DESCRIBE DATAFRAME ### ")
stats = df.describe()

print('\n'," ### INFO DATAFRAME ### ")
df.info()

print('\n'," ### Unique vals DATAFRAME ### ")
unique_vals = df.nunique()

print('\n'," ### Metrics ### ")
metrics_income = {'sum':round(df['income'].sum(), 2) ,
                  'avg':round(df['income'].mean(), 2) ,
                  'std':round(df['income'].std(), 2)
                  }

print(metrics_income)
```

```python
"""
Fill values

- Nulls
- Drop Fills
- Handling fill values
    - mean
    - model : {knn / randomforest etc...}
    - drop

"""

print(" ## Verify nulls ##",'\n')
df_null_chek =  df.isnull() 
compile_null = df.isnull().sum()

print(" ## Drop Nulls ##",'\n')
df_notnull = df.dropna() #Destroy null values
compile_null2 = df_notnull.isnull().sum()

print(df['idx'].count(), df_notnull['idx'].count())

print(" ## Handle nulls ##",'\n')
df_fill1 = df.fillna(method = 'mean')
```

```python
"""
Saving Dataframes 
- Save csv
"""
df.to_csv(path + "new_world_00.csv")
```
